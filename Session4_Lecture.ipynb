## 2. Transform the content to a list of rules 

We know that each rule starts with **"SCOPE SENTENCE"** so that will be indicator of rule begining. The rules ends on the line before the next mention of **"SCOPE SENTENCE"**.


```python
#as output i want a list of rules
rules=[]

#as output i want to keep information about where the rule starts

lines=[]

#my container

single_rule=""

for i in range(len(content)):
 
  if "SCOPE SENTENCE" in content[i]:

    # i reached the end of a rule and thus i put in the output with the line number

    rules.append(single_rule)

    lines.append(i+1)

    single_rule=""

  else:
    #adding lines o my container
    
    single_rule=single_rule + content[i]
```

## 5. Clean my code by organizing everthing in methods

```python
import pandas as pd

def efe_to_csv(path: str)-> pd.DataFrame:

  # read the file in a list of lines 

  with open(path,'r',encoding='utf-8',
                 errors='ignore') as script:
   content = script.read().splitlines()

   # transfrom the list of lines onto a Data Frame of rules with their corresponding begining line in the original file

  rules=[]

  lines=[]

  single_rule=""

  for i in range(len(content)):
 
    if "SCOPE SENTENCE" in content[i]:

      rules.append(single_rule)

      lines.append(i+1)

      single_rule=""

    else:
    
      single_rule=single_rule+content[i]   

    #Save everthing in a data frame and return it as a result of this function 

  structured_content= {'Rule':rules,'Line':lines}

  my_rules=pd.DataFrame(structured_content)

  return my_rules
```

```python
import pandas as pd

def search_rules(rules, queries: list)-> pd.DataFrame:

  selection=rules

  for query in queries:

    selection = selection[selection['Rule'].str.contains(query, regex=False)]

  return selection

```

```python
csv_rules= efe_to_csv("/content/drive/MyDrive/Colab Notebooks/11_features.efe")

csv_rules
```

```python
queries=['@T_11_Features[KEYWORD(EXPAND "features.txt")]', '@S_Negative[KEYWORD(EXPAND "negative_adj.txt")]']

search_rules(csv_rules, queries)
```
